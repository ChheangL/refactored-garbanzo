```python
def cvs_to_object(width,height,angle):
        data_frame = pd.read_csv((str(width)+'x'+str(height)+'_'+str(angle)+'frame.cvs'))
        data = {}
        for key in data_frame:
            data[key] = data_frame[key][0].replace('[','').replace(']','').split('\n')
            
            array = np.empty((0,2),float)
            for item in data[key]:
                values=item.split(' ')
                if values[-2] != '' and  values[-1]!= '':
                    array = np.append(array,[[float(values[-2]), float(values[-1])]],axis=0)
                data[key] = array
        return data
```
